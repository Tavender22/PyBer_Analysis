```python
# Import Depencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# skill test
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="*", color="green", linewidth=2, label='Boston')
ax.set_xlabel("Date")
ax.set_ylabel("Fares($)")
ax.set_title("PyBer Fare by Month")
ax.legend()
ax.set_ylim(0, 45)
ax.grid()
```

```python
# scatter plots
plt.plot(x_axis, y_axis, 'o')
```

```python
# scatter plots
plt.scatter(y_axis, x_axis, color="red", label="Chicago")
plt.title("Fare by Month($)")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.legend()
plt.gca().invert_yaxis()
```

```python
# bubble plots
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(y_axis, x_axis, s=y_axis_larger, color="red", label="Chicago")
plt.title("Fare by Month($)")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.legend()
plt.gca().invert_yaxis()
```

```python
# bubble plots
plt.scatter(y_axis, x_axis, s = [i*3 for i in y_axis], color="red", label="Chicago")
plt.title("Fare by Month($)")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.legend()
plt.gca().invert_yaxis()
```
